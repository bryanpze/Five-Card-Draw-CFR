```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
against_simplified = pd.read_csv('./simulation_results/Average Payoff of ExternalSampling No Suit - ExternalSampling No Suit V Article.csv',names=['Payoff Against Article'])
against_call = pd.read_csv('./simulation_results/Average Payoff of ExternalSampling No Suit - ExternalSampling No Suit V Call.csv',names=['Payoff Against Call'])
against_random = pd.read_csv('./simulation_results/Average Payoff of ExternalSampling No Suit - ExternalSampling No Suit V Random.csv',names=['Payoff Against Random'])
against_raise = pd.read_csv('./simulation_results/Average Payoff of ExternalSampling No Suit - ExternalSampling No Suit V Raise.csv',names=['Payoff Against Raise'])
```

```python
df = pd.DataFrame({
    "against_article": against_simplified['Payoff Against Article'],
    "against_call": against_call['Payoff Against Call'],
    "against_random": against_random['Payoff Against Random'],
    "against_raise": against_raise['Payoff Against Raise']
})

df_cumsum = df.cumsum()

df_cumsum.plot()
plt.xlabel("Number of Iterations")
plt.ylabel("Cumulative Payoffs")
plt.title("Cumulative Payoffs Against Weak Bechmarks")
plt.legend(loc="upper left")
plt.savefig("cum_payoff_weak.png")
plt.show()
```

```python
print(stats.ttest_1samp(against_simplified, popmean = 0,alternative='greater'))
print(stats.ttest_1samp(against_call, popmean = 0,alternative='greater'))
print(stats.ttest_1samp(against_random, popmean = 0,alternative='greater'))
print(stats.ttest_1samp(against_raise, popmean = 0,alternative='greater'))
```

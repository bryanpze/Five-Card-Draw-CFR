```python
big_b_highest_v_highest = pd.read_csv('./simulation_results/Big Blind Payoff of ExternalSampling Highest Suit Mini - ExternalSampling Highest Suit Mini V ExternalSampling Highest Suit Mini.csv',names=['Big Blind Payoff To Highest Suit Mini'])
big_b_no_suit_v_no_suit = pd.read_csv('./simulation_results/Big Blind Payoff of ExternalSampling No Suit - ExternalSampling No Suit V ExternalSampling No Suit.csv',names=['Big Blind Payoff To No Suit'])
small_b_highest_v_highest = pd.read_csv('./simulation_results/Small Blind Payoff of ExternalSampling Highest Suit Mini - ExternalSampling Highest Suit Mini V ExternalSampling Highest Suit Mini.csv',names=['Small Blind Payoff To Highest Suit Mini'])
small_b_no_suit_v_no_suit = pd.read_csv('./simulation_results/Small Blind Payoff of ExternalSampling No Suit - ExternalSampling No Suit V ExternalSampling No Suit.csv',names=['Small Blind Payoff To No Suit'])
```

```python
df = pd.DataFrame({
    "big_b_highest_v_highest": big_b_highest_v_highest['Big Blind Payoff To Highest Suit Mini'],
    "big_b_no_suit_v_no_suit": big_b_no_suit_v_no_suit['Big Blind Payoff To No Suit'],
    "small_b_highest_v_highest": small_b_highest_v_highest['Small Blind Payoff To Highest Suit Mini'],
    "small_b_no_suit_v_no_suit": small_b_no_suit_v_no_suit['Small Blind Payoff To No Suit']
})

df_cumsum = df.cumsum()

# Plot all columns in one single plot
df_cumsum.plot()
plt.xlabel("Number of Iterations")
plt.ylabel("Cumulative Payoffs")
plt.title("Cumulative Payoffs of Small Blinds V Big Blinds")
plt.legend(loc="upper left")
plt.savefig("cum_payoff_blinds.png")
plt.show()
```
